```python
import xarray as xr
import numpy as np
import os
```

```python
steric2 = xr.open_zarr('/swot/SUM02/spencer/steric_inc_mean.zarr')
steric_mean = xr.open_zarr('/swot/SUM02/spencer/swot_agulhas_surf/steric_mean.zarr')
```

```python
steric2 = steric2.steric - steric_mean.steric
```

```python
steric2.isel(time=20).plot()
```

```python
outdir = '/swot/SUM01/spencer/swot_agulhas_T_S/'
def output_binary_file(data, fname, dtype=np.dtype('>f4')):
    data.astype(dtype).tofile(fname)
```

```python
outdir='/swot/SUM03/spencer/swot_steric/'
for i in range(0,1751):
    if np.mod(i,10)==0:
        print(i)
    nt = 10368 + i*144
    Nt = nt - 10368
    eta = steric2.isel(time=i).values
    output_binary_file(eta, os.path.join(outdir, 'steric.%010d.data' % Nt))
```
